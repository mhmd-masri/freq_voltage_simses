```python
from graphviz import Digraph
```

```python
# Create flowchart for FFR1
ffr1 = Digraph('FFR1', comment='Flowchart for FFR1')

# Add nodes for FFR1 logic
ffr1.node('A', 'Start')
ffr1.node('B', 'Frequency < 49.9 Hz?')
ffr1.node('C', 'Set t_start = time, Increment Counter')
ffr1.node('D', 'Time - t_start < 30s?')
ffr1.node('E', 'Calculate Power Output')
ffr1.node('F', 'Enter Recovery Phase (0 Power Output)')
ffr1.node('G', 'Reset Counter')

# Add edges for FFR1 logic
ffr1.edges(['AB', 'BC', 'CD', 'DE', 'EF', 'FG'])

# Render the flowchart to a file (optional)
ffr1.render('ffr1_flowchart', format='png')

# Display the flowchart
ffr1.view()
```
